
<h3 id="Example-(part-A):">Example (part A):<a class="anchor-link" href="#Example-(part-A):">¶</a></h3><p>Consider the following ODE $$\dfrac{d\ y}{d\ x}=2\ y,$$</p>
<p>This is a separable equation with general solution $y=C e^{2x}$ (We have seen this in class in week 1 and 2)</p>
<p>We can use python to verify that the solution is correct (see section 0.2.5 pg 12/13 if you want an extra example).</p>
<p>First we need to define the symbols <code>x</code> and <code>C</code>, and the general solution <code>y</code></p>


```

x,C = sp.symbols('x C') # define x anc C as symbols for python to interpret them

y = C * sp.exp(2*x) # general solution

y

```


<p><strong>Note</strong>:</p>
<ul>
<li><code>sp.exp(x)</code> = $e^x$</li>
<li><code>sp.sin(x)</code> = $\sin(x)$</li>
<li><code>x**2</code> = $x^2$</li>
<li><code>sp.Rational{1,2}</code> = $\dfrac{1}{2}$  (for fractions with constant terms)</li>
</ul>



<p>The command <code>sp.diff(y,x)</code> takes the derivative of <code>y</code> with respect to <code>x</code></p>


```

sp.diff(y,x)

```


<p><strong>To verify if the solution is correct</strong>, we can plug it in the ODE and check that we get a <code>0==0</code> situation (python requires <code>==</code> instead of <code>=</code> to verify equality). Note: Python can recognize if two expressions are equal, but might not recognize identities, for example it fails fail to recognize $$1-\sin^2 x = \cos^2 x$$</p>


```

sp.cos(x)**2 == 1 - sp.sin(x)**2 # python does not recognize identities

```


<p>So to verify the solution we will move all of the terms to the LHS and leave the RHS equal to zero. So we change the ODE to $$\dfrac{d\ y}{d\ x} - 2\ y=0,$$ and then we ask python to simplify the expression using <code>sp.simplify()</code></p>


```

sp.simplify( sp.diff(y,x) - 2*y ) == 0 # we verify y is solution of ODE

```


<p>Given some initial conditions:</p>
<ul>
<li>$y(0)=2\quad \rightarrow \quad C=2$</li>
<li>$y(1)=e^2\quad \rightarrow \quad C=1$</li>
<li>$y(0)=4\quad \rightarrow \quad C=4$</li>
</ul>



<p>To plot these particular solutions we first need to substitute the constant $C$ with the correct value</p>


```

afewsolutions = [y.subs({C:c}) for c in [2,1,4]] # substitute C to get particular solutions
afewsolutions

```

```

for solution in afewsolutions: # go through solutions
    expressionplot(solution, x, 0, 3) # (expression_to_plot, variable, domain_start, domain_end)
plt.grid() # add grid

```


<h2 id="Example-(part-B):">Example (part B):<a class="anchor-link" href="#Example-(part-B):">¶</a></h2>



<p>Consider the linear differential equation we have seen in class $$\frac{d\ y}{d\ x}=x^3 - 2xy$$</p>



<p>This equation can be solved using the integrating factor technique (Ch 1.4) to obtain the general solution $$y = \frac{1}{2}x^2 - \frac{1}{2} + \frac{C}{e^{x^2}}$$</p>



<p>We want to use python to produce the slope field associated to the ODE.</p>
<p>So first we define the RHS as a function <code>f</code> then we use the <code>slopefieldplot</code> functionality to plot the slope field (note see section 1.2.3 pg 31,32 for another example)</p>


```

def f(x,y) : return x**3 - 2*x*y # define function of (x,y)

slopefieldplot(f,-2,2,-1,3,.2,lw=2)
# slopefieldplot(function, xmin, xmax, ymin, ymax, how_close_to_put_ticks, line_width)
plt.xlabel('x') # add x label
plt.ylabel('y') # add y label
plt.show() # for better output

```


<p>In class given 2 initial conditions we computed the corresponding particular solution:</p>
<ul>
<li>$y(0)=-\dfrac{1}{2} \quad \rightarrow \quad y_1 = \dfrac{1}{2}x^2 - \dfrac{1}{2}$</li>
</ul>
<ul>
<li>$y(0)=2 \quad \rightarrow \quad y_2= \dfrac{1}{2}x^2 - \dfrac{1}{2} + \dfrac{5}{2e^{x^2}}$</li>
</ul>



<p>Let's define these two functions using the same commands as in example A (above)</p>


```

x = sp.symbols('x') # define x as a symbol

y1 = sp.Rational(1,2) * x**2 - sp.Rational(1,2) # define particular solution y1

y2 = sp.Rational(1,2) * x**2 - sp.Rational(1,2) + sp.Rational(5,2)/sp.exp(x**2) # define particular solution y2

y1,y2 # display solutions

```


<p>Now let's plot the two solutions on the slope field using <code>expressionplot()</code></p>


```

slopefieldplot(f,-2,2,-1,3,.2,lw=2) # plot slope field
expressionplot(y1,x,-2,2,color='g',alpha=0.5,lw=3) #  plot y1
expressionplot(y2,x,-2,2,color='b',alpha=0.5,lw=3) #  plot y2
# color: sets the color ('r','b','g','k')
# alpha: transparency (alpha=1 full color, alpha=0 invisible)
# lw: line width
plt.xlabel('x') # add x label
plt.ylabel('y') # add y label
plt.show() # for better output

```
