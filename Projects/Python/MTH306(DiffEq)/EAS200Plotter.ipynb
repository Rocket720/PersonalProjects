```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
voltage = np.arange(0, 20, 0.1)
current_1kOhm = []
current_10kOhm = []
current_1MOhm = []

for v in voltage:
    current_1kOhm.append(v/1000)
    current_10kOhm.append(v/10000)
    current_1MOhm.append(v/1000000)

plt.plot(voltage, current_1kOhm, label = "1 kΩ")
plt.plot(voltage, current_10kOhm, label = "10 kΩ")
plt.plot(voltage, current_1MOhm, label = "1 MΩ")
plt.xlabel("Voltage (V)")
plt.ylabel("Current (A)")
plt.savefig('demo.png', transparent=True)
plt.legend()
```
